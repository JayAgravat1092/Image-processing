```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C://Users//jayag//Desktop//Marwadi//SEM-6//CV//Building.jpg")
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray)
```

```python
img_gray_flatten = np.float32(img_gray)
```

```python
plt.imshow(img_gray_flatten)
```

```python
harris_img = cv2.cornerHarris(img_gray_flatten,2,3,0.4)
```

```python
dilate_img = cv2.dilate(harris_img, None)
```

```python
img[dilate_img > 0.01 * dilate_img.max()] = [0, 0, 255]
```

```python
plt.figure(figsize = [10, 10])
plt.axis('off')
plt.imshow(img[:,:,::-1])
```
